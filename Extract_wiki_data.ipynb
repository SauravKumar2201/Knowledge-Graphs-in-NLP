 ## **Extract sentences from a Wikipedia page and save them to a CSV file**

```python
# Import packages
import wikipedia
import csv

# Specify the title of the Wikipedia page
page_title = 'Virat Kohli'

# Fetch the content of the Wikipedia page
wiki = wikipedia.page('Virat_Kohli')

# Extract the plain text content of the page, excluding images, tables, and other data.
text = wiki.content

# Replace '==' with '' (an empty string)
text = text.replace('==', '')

# Split the text into sentences
sentences = text.split('. ')

# Output CSV file
csv_filename = f"{page_title}_sentences.csv"

# Writing sentences to a CSV file
with open(csv_filename, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Sentence"])
    for sentence in sentences:
        writer.writerow([sentence])
```
